```python
import cv2
import skimage as ski
import matplotlib.pyplot as alt
```

```python
img = cv2.imread('pngegg.png')
```

```python
type(img)
```

```python
img.shape
```

```python
# to show the image 
cv2.imshow('Image', img)

# to hold the image on the window until interrupt
cv2.waitKey(0)

# to close the image when done
cv2.destroyAllWindows()
```

```python
# loading an image using matplotlib
img1 = alt.imread('pngegg.png')
alt.imshow(img1)
```

```python
# showing an image using skimage
img2 = ski.io.imread('pngegg.png')
ski.io.imshow(img2)
```

```python
# saving an image in different file
ski.io.imsave('myimg.png',img)
alt.imsave('myimg1.png',img)
cv2.imwrite('myimg2.png',img)
```

```python
cv2.COLOR_RGB2GRAY
```

```python
# convert image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
ski.io.imshow(img_gray)
```

```python
_, img_binary = cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY)
ski.io.imshow(img_binary,)
```
