```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('pngegg_0.png')
img2 = cv2.imread('pngegg_1.png')
img2 = img2[150:150+1024,150:150+1024,:]
```

```python
img.shape,img2.shape
```

```python
def showing(image):
    plt.imshow(image[:,:,::-1])
    plt.show()

def imshow(image):
    cv2.imshow('Preview',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
imshow(img)
imshow(img2)
```

```python
showing(img)
showing(img2)
```

```python
added = cv2.add(img2,img)
showing(added)
```

```python
subtracted = cv2.subtract(img,img2)
showing(subtracted)
```

```python
import numpy as np
```

```python
# decrease the brightness by 2 times
img2_1 = cv2.multiply(img2, 2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
# increase the brightness by 2 times
img2_1 = cv2.subtract(img2, 2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
# control contrast (alpha) and brightness(beta)
img2_2 = cv2.convertScaleAbs(img,alpha=1.1,beta =7)
showing(img2_2)
```

```python
# increase or decrease the size of the image
img2_3 = cv2.resize(img2, (400,400), interpolation=cv2.INTER_CUBIC)
img2_3.shape
```

```python
showing(img2_3)
```

```python
# width(fx) , height(fy)
img2_4 = cv2.resize(img2,None, fx = 2, fy=0.5, interpolation=cv2.INTER_CUBIC)
img2_4.shape
```

```python
showing(img2_4)
```

```python
from skimage.transform import rotate
```

```python
# rotate an image using skimage, and let it resize accordingly
img2_5 = rotate(img2, angle=45, resize=True)
showing(img2_5)
```

```python
repeat = True
while repeat:
    for i in range(360):
        i = rotate(img2_3, angle = i, resize = True)
        cv2.imshow('',i)
        key = cv2.waitKey(1)
        if key == ord('q'):
            repeat = False
            break
cv2.destroyAllWindows()
```
